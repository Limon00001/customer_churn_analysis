#### Make Data more Understable

If we look the data closely we can see that, `SeniorCitizen's` values are 1 and 0. That means it represents *True* or *False*. If we create a normal function that represents that customer is senior citizen or not then it will be more useful.

```python
def check_citizen(value):
  if value == 1:
    return 'YES'
  else:
    return 'NO'

# Use function
df['SeniorCitizen'] = df['SeniorCitizen'].apply(check_citizen)
```

```python
# Count the values for the stacked bar chart
count_data = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)
'''
  * df.groupby(['SeniorCitizen', 'Churn']).size() is a way to group the data in the DataFrame df based on the values in the columns SeniorCitizen and Churn,
  and then count the occurrences of each combination of those groups. It look like this:
  SeniorCitizen	  Churn
  NO	            No	  4508
                  Yes   1393
  YES	            No	  666
                  Yes	  476

  * unstack(fill_value = 0) Converts the grouped data from a Series to a DataFrame format where the unique values of Churn become columns. Any missing values
  (where there are no occurrences for a combination of SeniorCitizen and Churn) are filled with 0.
'''

# Step 2: Calculate percentages
percentages = count_data.div(count_data.sum(axis=1), axis=0) * 100
'''
  * count_data.sum(axis=1): Calculates the sum of each row (for each SeniorCitizen status) to get the total count for that status.
  * .div(..., axis=0): Divides each count by the total count of its row, giving the percentage of each Churn status within that SeniorCitizen category.
'''

# Step 3: Create a stacked bar chart
ax = count_data.plot(kind='bar', stacked=True, figsize=(5, 5), color=['lightblue', 'salmon'])
'''
  plot(kind='bar', stacked=True): Creates a stacked bar chart. Each bar represents a category (SeniorCitizen), with segments showing the counts of Churn categories stacked on top of each other.
  figsize=(5, 5): Sets the size of the figure to 5 inches by 5 inches.
  color=['lightblue', 'salmon']: Specifies the colors for each segment in the stacked bars.
'''

# Step 4: Add labels
for i in range(len(count_data)):
    cumulative_height = 0
    for j in range(len(count_data.columns)):
        cumulative_height += count_data.iloc[i, j]
        ax.text(i, cumulative_height - count_data.iloc[i, j] / 2,
                f'{percentages.iloc[i, j]:.1f}%',
                ha='center', va='center', color='black')
'''
  * Outer Loop: Iterates through each row (each SeniorCitizen status).
  * cumulative_height = 0: Initializes a variable to track the cumulative height of the segments in the current bar.
  * Inner Loop: Iterates through each column (each Churn category) within that row.
  * cumulative_height += count_data.iloc[i, j]: Adds the height of the current segment to the cumulative height.
  * ax.text(...): Places a text label at the specified coordinates on the plot.
  * i: The x-coordinate (index for SeniorCitizen).
  * cumulative_height - count_data.iloc[i, j] / 2: The y-coordinate (middle of the current segment).
  * f'{percentages.iloc[i, j]:.1f}%': The text displayed, formatted to one decimal place.
  * ha='center', va='center': Horizontally and vertically centers the text within the segment.
  * color='black': Sets the text color to black.
'''

plt.title('Count of Customers by Senior Citizen Status')
plt.xlabel('Senior Citizen')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])
plt.legend(title='Churn', labels=['Churn Out', 'Churn'])
plt.show()
```

```python
# List of columns to create count plots for
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up the subplots
num_columns = len(columns)
fig, axes = plt.subplots(nrows = (num_columns + 2) // 3, ncols = 3, figsize = (15, 5 * ((num_columns + 2) // 3)))
axes = axes.flatten()

# Create count plots for each column
for i, column in enumerate(columns):
    sns.countplot(data = df, x = column, ax = axes[i], hue = df['Churn'])
    axes[i].set_title(f'Countplot of {column}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()



'''
Summary of this visualization:

The majority of customers who do not churn tend to services like 'PhoneService', 'InternetService' (particularly DSL) and 'OnlineSecurity' enabled. For services
like 'onlineBackup', 'TechSupport', 'StreamingTV' churn rates are noticeably  higher when these services are not used or are unavailable.
'''
```
